```python
from matplotlib import pyplot as plt
```

## Line Graph:

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14858.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.xlabel("Years")

plt.show()
```

```python
variance = [2**i for i in range(9)]
bias_squared = variance[::-1] #invertendo a lista utilizando slice
total_error = [x + y for x, y in zip(variance, bias_squared)] #descompactação das listas variance e bias squared em x e y
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9) #localização 9 é top center
plt.xlabel("Model Complexity")
plt.xticks([])
plt.title("The bias-variance Tradeoff")
plt.show()
```

## Bar Graph:

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(range(len(movies)), num_oscars)

plt.title("Oscar winning movies")

plt.ylabel("Nº of oscars")

plt.xticks(range(len(movies)), movies)
plt.show()
```

## Histogram (values distribution):

```python
from collections import Counter
import random
```

```python
grades = [random.randrange(100) for _ in range(15)]

histogram = Counter(min(grade // 10*10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()], #deslocando barras para direita em 5 
        histogram.values(),
        10, #largura da barra
        edgecolor=(0,0,0))
plt.axis([-5, 105, 0, 5]) #x axis -5 to 105, y axis 0 to 5

plt.xticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of Exam grades")
plt.show
```

```python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times i heard someone say 'datascience'")
plt.ticklabel_format(useOffset=False)

plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'huge' increase!")
plt.show()
```

## Scatter Plots:

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = [chr(i) for i in range(97, 106)]

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                 xy=(friend_count, minute_count), #label on respective point
                 xytext=(5,-5), #move it a little bit
                 textcoords='offset points')
plt.title("Daily Minutes vs Number of Friends")
plt.xlabel("# of Friends")
plt.ylabel("daily minutes on site")

plt.show()
```

```python
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes are comparable")
plt.xlabel("test 1 grades")
plt.ylabel("test 2 grades")
plt.axis("equal")
plt.show()
```
